```python
#3
from queue import Queue

def createGen2():
    for i in range(1, 20):
        yield i

gen = createGen2()

class Pump:
    
    def __init__(self, queue, gen):
        self.queue = queue
        self.gen = gen

    def action(self):
        self.queue.put(next(self.gen))

class MultiAction():
    
    def __init__(self, pump, n):
        self.pump = pump
        self.n = n

    def action(self):
        for _ in range(self.n):
            self.pump.action()

class MultiPump():

    def __init__(self, queues, gen):
        self.queues = queues
        self.gen = gen
        self.current_queue = 0

    def action(self):
        self.queues[self.current_queue].put(next(self.gen))
        self.current_queue = (self.current_queue + 1)%len(self.queues)
        
```

```python
queue = Queue()
pump = Pump(queue, createGen2())
m_action = MultiAction(pump, 6)
m_action.action()

while not queue.empty():
    print(queue.get())

queues = [Queue(), Queue(), Queue()]
m_pump = MultiPump(queues, createGen2())

for i in range(8):
    m_pump.action()

print("\n")

while not queues[0].empty():
    print(queues[0].get())
    
```

```python
#4
from queue import Queue
import random

def gen_message():
    text = ["test1", "test2", "test3", "test4", "test5"]
    while True:
        yield random.choice(text)

class MultiPump():

    def __init__(self, queues, gen):
        self.queues = queues
        self.gen = gen
        self.current_queue = 0

    def action(self):
        self.queues[self.current_queue].put(next(self.gen))
        self.current_queue = (self.current_queue + 1)%len(self.queues)

class GenFromQ():
    
    sharedQueue = Queue()
    
    def __init__(self, queues):
        self.queues = queues

    def action(self):
        m_pump = MultiPump(self.queues, gen_message())
        for i in range(10):
            m_pump.action()

        for queue in self.queues:
            while not queue.empty():
                self.sharedQueue.put(queue.get())

        
q = [Queue(), Queue(), Queue()]

gfq = GenFromQ(q)
gfq.action()


while not gfq.sharedQueue.empty():
    print(gfq.sharedQueue.get())
```
